```python
from ase.io import vasp
import pandas as pd
import re
import numpy as np
```

```python
response = []
for x in range(95):
    fname = '../simulation-data/sf300K_14Mn_Job1accurate/Run'+str(x)+'/OUTCAR'
    print(fname)
    last_match = None
    for line in open(fname):
        match = re.findall('energy  without entropy=    ([-+]?\d*\.\d+|\d+)', line)
        if match:
            last_match = match    
    print(last_match)
    response.append(float(last_match[0]))
for x in range(44):
    fname = '../simulation-data/bulk300K_7Mn_Job1accurate/Run'+str(x)+'/OUTCAR'
    print(fname)
    last_match = None
    for line in open(fname):
        match = re.findall('energy  without entropy=    ([-+]?\d*\.\d+|\d+)', line)
        if match:
            last_match = match
    print(last_match)
    response.append(float(last_match[0]))
for x in range(50):
    fname = '../simulation-data/sf300K_7Mn_Job1accurate/Run'+str(x)+'/OUTCAR'
    print(fname)
    last_match = None
    for line in open(fname):
        match = re.findall('energy  without entropy=    ([-+]?\d*\.\d+|\d+)', line)
        if match:
            last_match = match
    print(last_match)
    response.append(float(last_match[0]))
```

```python
response = np.array(response)
print(response.shape)
```

```python
molecules = []
for x in range(95):
    fname = '../simulation-data/sf300K_14Mn_Job1accurate/Run'+str(x)+'/CONTCAR'
    print(fname)
    molecule = vasp.read_vasp(file=fname)
    molecules.append(molecule)
for x in range(44):
    fname = '../simulation-data/bulk300K_7Mn_Job1accurate/Run'+str(x)+'/CONTCAR'
    print(fname)
    molecule = vasp.read_vasp(file=fname)
    molecules.append(molecule)
for x in range(50):
    fname = '../simulation-data/sf300K_7Mn_Job1accurate/Run'+str(x)+'/CONTCAR'
    print(fname)
    molecule = vasp.read_vasp(file=fname)
    molecules.append(molecule)
```

```python
print(len(molecules))
```

```python
structures = molecules
species = set()
for structure in structures:
    species.update(structure.get_chemical_symbols())
```

```python
# Let's configure the SOAP descriptor.
from dscribe.descriptors import SOAP

soap = SOAP(
    species=species,
    periodic=False,
    rcut=5,
    nmax=8,
    lmax=8,
    average="outer",
    sparse=False
)
```

```python
# Let's create SOAP feature vectors for each structure
feature_vectors = soap.create(molecules, n_jobs=8)
```

```python
feature_vectors.shape
```

```python
import numpy as np
dataset = np.concatenate([response.reshape(feature_vectors.shape[0],1), feature_vectors], axis=1)
```

```python
np.save('../training-data/all_data.npy', dataset)
```
